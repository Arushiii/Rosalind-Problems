```python
a = input('seq')
```

```python
l = []
for i in a:
    #print(i)
    if i=='A':
        #print('T')
        l.append('T')
    elif i=='T':
        #print('A')
        l.append('A')
    elif i=='G':
        #print('C')
        l.append('C')
    elif i=='C':
        #print('G')
        l.append('G')
    
```

```python
complement= ''.join(l)
```
